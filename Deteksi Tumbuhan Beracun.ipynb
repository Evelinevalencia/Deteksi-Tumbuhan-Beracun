```python
import cv2
import matplotlib.pyplot as plt

def extract_leaf_features(image):

    hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

    lower_green = (40, 40, 40)
    upper_green = (80, 255, 255)

    mask = cv2.inRange(hsv_image, lower_green, upper_green)

    edges = cv2.Canny(mask, 30, 150)

    leaf_area = cv2.countNonZero(edges)

    return leaf_area, edges

def classify_leaf(leaf_area):

    threshold = 5000
    if leaf_area > threshold:
        return 'Beracun'
    else:
        return 'Tidak Beracun'

image_path = 'download (2).jpg'
image = cv2.imread(image_path)

leaf_area, edges = extract_leaf_features(image)

classification = classify_leaf(leaf_area)
print(leaf_area)

plt.figure(figsize=(10, 5))

plt.subplot(121)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Gambar Asli')
plt.axis('off')

plt.subplot(122)
plt.imshow(edges, cmap='gray')
plt.title(f'Klasifikasi: {classification}')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

def segment_green(image):
    hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

    lower_green = (40, 40, 40)
    upper_green = (80, 255, 255)

    mask = cv2.inRange(hsv_image, lower_green, upper_green)

    result = cv2.bitwise_and(image, image, mask=mask)

    return result

def classify_segmentation(segmented_image):
    nonzero_pixels = cv2.countNonZero(cv2.cvtColor(segmented_image, cv2.COLOR_BGR2GRAY))
    threshold = 50000
    threshold1 = 300000
    if nonzero_pixels < threshold or nonzero_pixels > threshold1:
        return 'Beracun'
    else:
        return 'Tidak Beracun'

image_paths = ['download.jpg']

plt.figure(figsize=(15, 5))

for i, path in enumerate(image_paths, 1):
    image = cv2.imread(path)
    segmented_image = segment_green(image)
    classification = classify_segmentation(segmented_image)

    plt.subplot(1, len(image_paths), i)
    plt.imshow(cv2.cvtColor(segmented_image, cv2.COLOR_BGR2RGB))
    plt.title(f'{classification}')
    plt.axis('off')

plt.tight_layout()
plt.show()
```
